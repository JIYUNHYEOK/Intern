```python
data_residence = data[data['업태구분명']=='숙박업(생활)']
data_residence['label'] = '생활형숙박시설'
print(data_residence.shape)
data_residence.head()
```

```python
data_normal = data[data['업태구분명']!='숙박업(생활)']
data_normal['label'] = '일반숙박시설'
print(data_normal.shape)
data_normal.head()
```

```python
plt.figure(figsize=(18,6))
sns.countplot(x='year', data=data_residence, palette='Set2')
plt.title('생활형 숙박시설 증가추이', fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.countplot(x='year', data=data_normal, palette='Set2')
plt.title('일반 숙박시설 증가추이', fontsize=25)
plt.show()
```

```python
data_all = pd.concat([data_normal, data_residence])
```

```python
# 일반숙박업 증가 수
list_normal = list(data_normal.groupby(['year'])[['label']].count()['label'])

# 생활혈숙박업 증가 수
list_residence = list(data_residence.groupby(['year'])[['label']].count()['label'])
```

```python
fig, ax = plt.subplots(figsize=(18,6))
bar_width = 0.4

index = np.arange(12)
year = np.arange(2010,2022,1)

b1 = plt.bar(index, list_normal, bar_width, alpha=0.4, color='red', label='일반숙박시설')

b2 = plt.bar(index + bar_width, list_residence, bar_width, alpha=0.4, color='blue', label='생활형숙박시설')

plt.xticks(np.arange(0.2, 12 , 1), year)

plt.xlabel('year', size = 15)
plt.ylabel('count', size = 15)
plt.title('일반숙박시설과 생활형숙박시설 증가 비교', fontsize = 20)
plt.legend(fontsize = 20)
plt.show()
```
